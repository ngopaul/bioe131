```python
from matplotlib import pyplot as plt
import os
```

## Create a cleaning function for the words, and add all the words in the file to the dictionary

```python
def clean(word):
    return ''.join(l for l in word.lower() if 'a' <= l <= 'z')

d = {}

with open("romeoandjuliet.txt") as f:
    for line in f:
        for word in line.split():
            word = clean(word)
            if word in d:
                d[word] += 1
            else:
                d[word] = 1
```

## Sort the dictionary by using a list

```python
d
```

```python
l = list(d.items())

l.sort(key=lambda x:-x[1])

sorted_dict = dict(l)
```

```python
sorted_dict
```

## Obtain the ranks of the words instead of the words themselves

```python
x = list(sorted_dict.keys())
y = list(sorted_dict.values())
rank = [i for i in range(len(x))]
```

### Plot the rank vs the number of occurences

```python
fig, ax = plt.subplots()
plt.bar(rank, y)
ax.set_xlabel('rank of word')
ax.set_ylabel('number of occurences')
plt.show()
```

### Plot the rank vs the number of occurences on a log-log plot

```python
fig, ax = plt.subplots()
ax.set_yscale('log')
ax.set_xscale('log')
ax.set_xlabel('rank of word')
ax.set_ylabel('number of occurences')
plt.plot(rank, y)
plt.show()
```

## Interpretation

I think, since the plot looks almost linear for the majority of the words in the log-log plot, that means that in the english language, we use words that fit to a power fit. So we use a certain word 1/2 the time, another word 1/3 the time, 1/4, et cetera (multiplied by some constant).
